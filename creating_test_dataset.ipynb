### Loading emotion data 

```python
testing_emo_ = pd.read_csv('data/MotionEmotion/test_emo_dataset.csv')
testing_emo_
```

```python
nothing = testing_emo_[testing_emo_['test_labels']== 0].index
nothing
testing_emo_.drop(nothing , inplace=True)
testing_emo_
```

```python
test_seq_1 = testing_emo_[testing_emo_['test_labels']==1].sample(n=200)
test_seq_2 = testing_emo_[testing_emo_['test_labels']==2].sample(n=200)
test_seq_3 = testing_emo_[testing_emo_['test_labels']==3].sample(n=200)
test_seq_4 = testing_emo_[testing_emo_['test_labels']==4].sample(n=200)
test_seq_5 = testing_emo_[testing_emo_['test_labels']==5].sample(n=200)
test_seq_6 = testing_emo_[testing_emo_['test_labels']==6].sample(n=200)
test_set = pd.concat((test_seq_1,test_seq_2,test_seq_3,test_seq_4,test_seq_5,test_seq_6))
test_set
```

### Exporting reduced emotion test set 

```python
test_set.to_csv('data/MotionEmotion/test_set.csv', encoding='utf-8')
```
